## GDP per country (WORLD BANK GROUP API)

```python
path2 = "Data Frames/GDP 2022.xlsx"
df_gdp = pd.read_excel(path2)
pd.set_option('display.max_columns', None)
df_gdp
```

```python
gdp_percapta = df_gdp.drop(columns=['#', 'GDP\n(nominal, 2022)', 'GDP\n(abbrev.)',	'GDP\ngrowth', 'Share of\nWorld GDP'])
gdp_percapta
```

```python
gdp_percapta.rename(columns= {'GDP\nper capita': 'GDP per capta'}, inplace=True)
gdp_percapta.rename(columns= {'Population\n(2022)': 'Population'}, inplace=True)
```

```python
# Remove the $ and , characters and convert to int
gdp_percapta['GDP per capta'] = gdp_percapta['GDP per capta'].str.replace('$', '').str.replace(',', '').astype(int)
gdp_percapta
```

```python
new_order = ["Country", "GDP per capta", "Population"]
gdp_percapta = gdp_percapta[new_order]
gdp_percapta
```

```python
countries_ai = df_ai['Country'].unique()
countries_ai
```
