```python
import pandas as pd
import matplotlib.pyplot as plt


# Read the CSV file into a DataFrame with semicolon delimiter
df = pd.read_csv('tables/country_talent_gs_mostcrowdet_toless.csv', delimiter=';')

# Display the DataFrame
print(df)
```

```python
df
```

```python
countries = df['country']
populations = df['population']   # Convert population to millions
talents = df['talent']

# Create a bubble chart
plt.figure(figsize=(10, 6))
scatter = plt.scatter(populations, talents, s=populations / 1e6, alpha=0.4, c=talents, cmap='viridis', edgecolors='w', linewidth=0.5)

# Add labels and title
plt.xlabel('Population')
plt.ylabel('Talent')
plt.title('Population vs Talent by Country ')

# Add color bar to indicate talent score
plt.colorbar(scatter, label='Talent')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
populations = df['population'] / 1e6  # Convert population to millions
talents = df['talent']

# Create a bubble chart
plt.figure(figsize=(12, 8))
scatter = plt.scatter(populations, talents, s=df['population'] / 1e6, alpha=0.5, c=talents, cmap='viridis', edgecolors='w', linewidth=0.5)

# Add labels and title
plt.xlabel('Population (in millions)')
plt.ylabel('Talent')
plt.title('Population vs Talent by Country')

# Add color bar to indicate talent score
plt.colorbar(scatter, label='Talent')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
df2= pd.read_csv('tables/high_gs_low_talent.csv', delimiter=',')
```

```python
df2
```

```python


# Plotting talent scores with vertical country names
plt.figure(figsize=(10, 8))

# Scatter plot
plt.scatter(df2['Country'], df2['talent'], color='blue', marker='o', s=100)

# Set x-axis labels vertically
plt.xticks(rotation=90)

# Labels and title
plt.xlabel('Country')
plt.ylabel('Talent Score')
plt.title('Talent Scores by Country')

# Display plot
plt.tight_layout()
plt.show()
```

```python


# Plotting talent scores and government strategy
plt.figure(figsize=(12, 8))

# Scatter plot for talent scores
plt.scatter(df2['Country'], df2['talent'], color='blue', marker='o', s=100, label='Talent Score')

# Scatter plot for government strategy
plt.scatter(df2['Country'], df2['government_strategy'], color='red', marker='o', s=100, label='Government Strategy')

# Display government strategy values on the right side
for i, (talent, gov_strategy) in enumerate(zip(df2['talent'], df2['government_strategy'])):
    plt.text(i, gov_strategy, f"{gov_strategy:.2f}", va='center', ha='left', fontsize=10, color='red')

# Set x-axis labels vertically
plt.xticks(rotation=90)

# Labels and title
plt.xlabel('Country')
plt.ylabel('Score')
plt.title('Talent Scores and Government Strategy by Country')
plt.legend()

# Display plot
plt.tight_layout()
plt.show()
```

```python
# Plotting talent scores and government strategy
plt.figure(figsize=(12, 8))

# Scatter plot for talent scores
plt.scatter(df2['Country'], df2['talent'], color='blue', marker='o', s=100, label='Talent Score')

# Scatter plot for government strategy
plt.scatter(df2['Country'], df2['government_strategy'], color='red', marker='o', s=100, label='Government Strategy')

# Display government strategy values on the right side
for i, (talent, gov_strategy) in enumerate(zip(df2['talent'], df2['government_strategy'])):
    plt.text(i, gov_strategy, f"{gov_strategy:.2f}", va='center', ha='left', fontsize=9, color='red')

# Set x-axis labels vertically
plt.xticks(rotation=90)

# Labels and title
plt.xlabel('Country')
plt.ylabel('Score')
plt.title('Talent Scores and Government Strategy by Country')

# Create a legend for the plot
blue_marker = plt.Line2D([], [], color='blue', marker='o', markersize=10, linestyle='None', label='Talent Score')
red_marker = plt.Line2D([], [], color='red', marker='o', markersize=10, linestyle='None', label='Government Strategy')
plt.legend(handles=[blue_marker, red_marker], loc='upper left')

# Display plot
plt.tight_layout()
plt.show()
```

```python

fig, ax1 = plt.subplots(figsize=(12, 8))

# Scatter plot for talent scores
ax1.scatter(df2['Country'], df2['talent'], color='blue', marker='o', s=100, label='Talent Score')

# Scatter plot for government strategy
ax2 = ax1.twinx()
ax2.scatter(df2['Country'], df2['government_strategy'], color='red', marker='o', s=100, label='Government Strategy')

# Set y-axis limit for ax2 to start from 0
ax2.set_ylim(0, max(df2['government_strategy']) * 1.1)  # Adjust multiplier as needed for padding

# Set x-axis labels vertically
ax1.set_xticklabels(df2['Country'], rotation=90)

# Labels and title
ax1.set_xlabel('Country')
ax1.set_ylabel('Talent Score', color='blue')
ax2.set_ylabel('Government Strategy', color='red')
plt.title('Talent Scores and Government Strategy by Country')

# Create a legend for the plot
blue_marker = plt.Line2D([], [], color='blue', marker='o', markersize=10, linestyle='None', label='Talent Score')
red_marker = plt.Line2D([], [], color='red', marker='o', markersize=10, linestyle='None', label='Government Strategy')
plt.legend(handles=[blue_marker, red_marker], loc='upper left')

# Display plot
plt.tight_layout()
plt.show()
```
